```python
# Import packages
import pandas as pd
import csv, json
```

```python
# Data file paths
nodespath = '../Data/personsummary-test.csv'
edgespath = '../Data/alledge.csv'
netjson = '../Visualization/allnetwork-test.json'
```

```python
# Import node data
nodes = pd.read_csv('../Data/personsummary-test.csv')
nodes
```

```python
# Import edge data
edges = pd.read_csv('../Data/alledge.csv')
edges
```

```python
# Initialize JSON data
jsondata = {"nodes":[], "links": []}
jsondata
```

```python
# Iterate through nodes CSV and add relevant fields to JSON input data
cnt = 0

with open(nodespath) as nodes:
    csvReader = csv.reader(nodes)
    for row in csvReader:
        if cnt != 0:
            # Create node dictionary
            node = {}
            node['id'] = row[0]
            node['totnumenc'] = row[1]
            node['totsec'] = row[3]
            node['img'] = row[4]
            # Append to json data
            jsondata['nodes'].append(node)
        cnt += 1
        
jsondata
```

```python
# Iterate through edges CSV and add relevant fields to JSON input data
cnt = 0

with open(edgespath) as edges:
    csvReader = csv.reader(edges)
    for row in csvReader:
        if cnt != 0:
            # Create edge dictionary
            edge = {}
            edge['source'] = row[0]
            edge['target'] = row[1]
            edge['value'] = row[4]
            # Append to json data
            jsondata['links'].append(edge)
        cnt += 1
        
jsondata
```

```python
with open(netjson, 'w') as jsonFile:
    jsonFile.write(json.dumps(jsondata, indent=4))
```
